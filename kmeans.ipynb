```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
```

```python
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1]);
```

1. Procéder pour classer les données à l'aide de l'algorithme K-Means avec python.

```python
kmeans = KMeans(n_clusters=4, random_state=0).fit(X)
kmeans.labels_
```

```python
pred = kmeans.predict(X)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(X[:, 0], X[:, 1], c=pred)
plt.show()
```

2. Déterminer le nombre optimal de cluster en utilisant la Elbow method:
Comme mentionné avant, nous  essayons plusieurs modèles en utilisant un nombre différent de clusters et en stockant la valeur (WCSS) à chaque fois.

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit_predict(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

3. Classer les données en utilisant le nombre optimal de clusters déterminé à la dernière étape.


```python
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(X)
plt.scatter(X[:,0], X[:,1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.show()
```

## Comparer Kmeans avec et sans Elbow

```python
# générer une séquence aléatoire

x1 = 10*np.random.rand(100,2)
x1.shape
```

```python
# plot x1

plt.scatter(x1[:, 0], x1[:, 1]);
```

```python
# essayer avec trois clusters

kmeans = KMeans(n_clusters=3)
pred_x1 = kmeans.fit_predict(x1)
plt.scatter(x1[:,0], x1[:,1], c=pred_x1)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.show()
```

```python
# procéder avec la méthode d'Elbow

wcss = []
for i in range(1, 21):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit_predict(x1)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 21), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# choix du nombre de clusters: 10

kmeans = KMeans(n_clusters=10, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_x1 = kmeans.fit_predict(x1)
plt.scatter(x1[:,0], x1[:,1], c=pred_x1 )
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.show()
```
